```python
'''
This will display the summary statistics of all of the variables
'''
import numpy as np
import pandas
from matplotlib import pyplot as plt
data = pandas.read_csv('csv/full_image_metadata.csv')
data.describe(include='all')
```

```python
'''
This will display the summary statistics of the numerical statistics that have 
any variance by removing those with a standard deviation of zero
'''
# filtering out the columns with no standard deviation
numeric_data = data.select_dtypes(include=[np.number])
std_dev = numeric_data.std()
(std_dev[std_dev > 0].index.tolist())
columns_to_include = (std_dev[std_dev > 0.000000001].index.tolist())
data[columns_to_include].describe()
```

```python
# try plotting some of the variables on a histagram
plt.hist(data['ISOSpeedRatings'])
plt.show
plt.close()
plt.hist(data['ExposureTime'])
plt.show
plt.close()
figure, axis = plt.subplots(3, 3)
axis[0,0].hist(data['Latitude'])
axis[0,1].hist(data['Longitude'])
axis[0,2].hist(data['ShutterSpeedValue'])
axis[1,0].hist(data['ExposureBiasValue'])
axis[1,1].hist(data['MeteringMode'])
axis[1,2].hist(data['LightSource'])
axis[2,0].hist(data['ExposureTime'])
axis[2,1].hist(data['ISOSpeedRatings'])
figure.delaxes(axis[2,2])
plt.tight_layout()
```

```python
'''
This will display the summary statistics of the categorical statistics
'''
categorical_data = data.select_dtypes(exclude=[np.number])
print(categorical_data['Model'])
categorical_data.describe()
```
